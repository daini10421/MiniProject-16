# **"Resize the 100 images to 50% (smaller) " using OpenCV**

```python
import cv2
import os
import requests
from urllib.parse import urlparse

# GitHub repository URL
repo_url = "https://github.com/daini10421/100images"

# Iterate over 100 images
for i in range(1, 101):
    # Create the URL to the raw image file
    image_url = f"{repo_url}/raw/main/image({i}).png"

    # Parse the URL
    parsed_url = urlparse(image_url)

    # Fetch the image filename from the URL
    image_filename = os.path.basename(parsed_url.path)

    try:
        # Download the image using requests
        response = requests.get(image_url)
        with open(image_filename, 'wb') as f:
            f.write(response.content)

        # Load the input image
        img = cv2.imread(image_filename)

        if img is not None:
            # Resize the image to 50% of its original size
            resized_img = cv2.resize(img, None, fx=0.5, fy=0.5)

            # Show the resized image
            cv2.imshow('Resized Image', resized_img)
            cv2.waitKey(0)

            # Remove the downloaded image
            os.remove(image_filename)
        else:
            print(f"Resized {image_filename} to 50%")

    except Exception as e:
        print(f"Error downloading image: {image_url}")
        print(f"Error message: {str(e)}")

# Close all windows
cv2.destroyAllWindows()
```
